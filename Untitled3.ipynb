```python
# Importamos las librerías necesarias para el procesamiento, modelado y gráficos
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score,
    recall_score,
    confusion_matrix,
    classification_report
)

# 1. Definición de la URL de tu repositorio de GitHub
url_github = "https://raw.githubusercontent.com/AlexTheco256/ProyectoAgenteInteligente/main/Dataset.csv"

print("Conectando con el repositorio")
# 2. Lectura inicial del archivo CSV con el separador correcto (coma)
df_original = pd.read_csv(url_github, sep=",", encoding="latin1")

print(f"El dataset original contiene {df_original.shape[0]} filas y {df_original.shape[1]} columnas.")
```

```python
print("\n--- VISTA PREVIA DE LAS 5 PRIMERAS FILAS DE LA DATA ORIGINAL ---")
# Mostramos una tabla interactiva con las primeras 5 filas en Colab
df_original.head()
```

```python
# 1. Definición de variables permitidas para evitar la filtración de datos (Data Leakage)
variables_numericas = ["AREA_TH", "AREA_PMF", "NUM_ARBOLES_APROBADOS", "CANT_ESPECIES_APROBADOS", "VOL_APROBADO"]
variables_categoricas = ["DEPARTAMENTO", "MODALIDAD_APROVECHAMIENTO"]
objetivo = "ESTADO_OBSERVATORIO"

# 2. Creamos una copia con las columnas seleccionadas
df_filtrado = df_original[variables_numericas + variables_categoricas + [objetivo]].copy()

# 3. Limpieza de formatos numéricos (conversión limpia a números)
for columna in variables_numericas:
    df_filtrado[columna] = df_filtrado[columna].astype(str).str.replace(",", "", regex=False).str.strip()
    df_filtrado[columna] = pd.to_numeric(df_filtrado[columna], errors='coerce')

# 4. Eliminación de registros con valores vacíos o inconsistentes
df_filtrado = df_filtrado.dropna()

print(f"Limpieza completada. Registros válidos para el agente: {df_filtrado.shape[0]} filas.")
print("\n--- RESUMEN ESTADÍSTICO DE LAS VARIABLES NUMÉRICAS SELECCIONADAS ---")

# Mostramos una tabla con métricas clave (promedio, desviación, min, max)
df_filtrado[variables_numericas].describe()
```

```python
# 1. Mapeo de la variable objetivo (0 = Verde, 1 = Roja)
y = df_filtrado[objetivo].map({"LISTA VERDE": 0, "LISTA ROJA": 1})

# 2. Aislamos las características de entrada descartando la respuesta
X_previa = df_filtrado.drop(columns=[objetivo])

# 3. Transformación de categorías de texto a columnas numéricas (One-Hot Encoding)
X = pd.get_dummies(X_previa, columns=variables_categoricas, drop_first=True)

print("--- ESTRUCTURA FINAL DE LAS VARIABLES DE ENTRADA (X) COMPUTADAS ---")
print(f"Matriz procesada para el agente: {X.shape[0]} registros listos con {X.shape[1]} características numéricas.")
print("\nColumnas codificadas finales (Vista de las primeras 5 filas):")

# Mostramos cómo lucen ahora las variables tras la transformación digital
X.head()
```

```python
# 1. Definición del Árbol de Decisión
def entrenar_arbol(X_train, y_train):
    # max_depth=5 evita que el árbol memorice la data (overfitting) y mantenga reglas lógicas simples
    modelo = DecisionTreeClassifier(
        max_depth=5,
        random_state=42
    )
    modelo.fit(X_train, y_train)
    return modelo

# 2. Definición del Bosque Aleatorio (Random Forest)
def entrenar_random_forest(X_train, y_train):
    # Creamos un ensamble de 100 árboles independientes que votarán por el resultado final
    modelo = RandomForestClassifier(
        n_estimators=100,
        random_state=42
    )
    modelo.fit(X_train, y_train)
    return modelo

print("Funciones de entrenamiento configuradas y listas para procesar.")
```
