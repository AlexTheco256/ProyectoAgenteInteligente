```python
# =========================================================================
# BLOQUE 1: PREPARACIÓN DE LOS DATOS (Dataset OSINFOR)
# =========================================================================
#!pip install pandas numpy matplotlib seaborn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("[INFO] Cargando registros originales de OSINFOR...")
df = pd.read_csv("https://raw.githubusercontent.com/AlexTheco256/ProyectoAgenteInteligente/main/Datasets.csv")

# 1.1 Filtrado de características críticas (Remoción de variables que causan Data Leakage)
# Evitamos usar columnas de fiscalización ex-post como 'VOLUMEN_ILEGAL' o 'ARBOLES_INEXISTENTES'
features_validas = [
    'MODALIDAD_APROVECHAMIENTO', 'DEPARTAMENTO', 'AREA_TH', 'AREA_PMF',
    'NUM_ARBOLES_APROBADOS', 'CANT_ESPECIES_APROBADOS', 'VOL_APROBADO', 'ESTADO_OBSERVATORIO'
]
df_limpio = df[features_validas].copy()

# 1.2 Tratamiento de valores nulos (Imputación por mediana para evitar sesgos en la escala)
for col in df_limpio.select_dtypes(include=[np.number]).columns:
    df_limpio[col] = df_limpio[col].fillna(df_limpio[col].median())

# 1.3 Mapeo y codificación de variables categóricas (One-Hot Encoding)
X_raw = df_limpio.drop(columns=['ESTADO_OBSERVATORIO'])
y = df_limpio['ESTADO_OBSERVATORIO'].map({'LISTA VERDE': 0, 'LISTA ROJA': 1})

# Codificación dummy de regiones y modalidades
X = pd.get_dummies(X_raw, columns=['DEPARTAMENTO', 'MODALIDAD_APROVECHAMIENTO'], drop_first=True)

print(f"[ÉXITO] Matriz de características lista. Dimensiones: {X.shape}")
```

```python
# =========================================================================
# BLOQUE 2: DIVISIÓN DEL DATASET Y MODELO BASELINE (EX-ANTE)
# =========================================================================
#!pip install scikit-learn
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report

# 2.1 Partición balanceada y estratificada (75% Entrenamiento / 25% Test)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, stratify=y, random_state=42
)

# 2.2 Modelo Baseline: Árbol de Decisión Libre (Sin control de profundidad)
print("[PROCESO] Entrenando modelo de control (Pre-Implementación)...")
modelo_pre = DecisionTreeClassifier(max_depth=None, random_state=42)
modelo_pre.fit(X_train, y_train)

# Evaluación inicial
y_pred_pre = modelo_pre.predict(X_test)
reporte_pre = classification_report(y_test, y_pred_pre, target_names=["LISTA VERDE", "LISTA ROJA"], output_dict=True)
print("\n--- REPORTE EX-ANTE (ÁRBOL SIMPLE) ---")
print(classification_report(y_test, y_pred_pre, target_names=["LISTA VERDE", "LISTA ROJA"]))
```

```python
# =========================================================================
# BLOQUE 3 Y 4: CONFIGURACIÓN, HIPERPARÁMETROS Y GESTIÓN DE DESBALANCEO
# =========================================================================
from sklearn.ensemble import RandomForestClassifier

print("[PROCESO] Configurando Agente Inteligente Regularizado (Post-Implementación)...")

# Inicialización con parámetros que controlan la varianza (Evita memorización/Overfitting)
modelo_post = RandomForestClassifier(
    n_estimators=300,            # Robustez estadística mediante número de estimadores
    max_depth=12,                # Límite de profundidad estricto contra Overfitting
    min_samples_split=6,         # Mínimo de muestras para crear ramas nuevas
    min_samples_leaf=4,          # Estabilidad de observaciones mínimas en nodos terminales
    class_weight='balanced',     # Solución directa al desbalanceo (Bloque 4)
    max_features='sqrt',         # Selección aleatoria de variables por árbol
    random_state=42,
    n_jobs=-1
)

# Entrenamiento del ensamble optimizado
modelo_post.fit(X_train, y_train)
print("[ÉXITO] Modelo Random Forest entrenado bajo restricciones matemáticas.")
```

```python
# =========================================================================
# BLOQUE 5: EVALUACIÓN CORRECTA DEL MODELO Y EXPORTACIÓN DE GRÁFICOS
# =========================================================================
from sklearn.metrics import confusion_matrix

# 5.1 Predicción final
y_pred_post = modelo_post.predict(X_test)
reporte_post = classification_report(y_test, y_pred_post, target_names=["LISTA VERDE", "LISTA ROJA"], output_dict=True)

print("\n--- REPORTE EX-POST (RANDOM FOREST OPTIMIZADO) ---")
print(classification_report(y_test, y_pred_post, target_names=["LISTA VERDE", "LISTA ROJA"]))
```

```python
# 5.3 GRÁFICO 2: Comparativa de Rendimiento F1-Score (Antes vs Después)
dicc_metricas = {
    'Modelo Evaluado': ['Pre-Implementación (Árbol)', 'Pre-Implementación (Árbol)',
                        'Post-Implementación (Random Forest)', 'Post-Implementación (Random Forest)'],
    'Clase de Alerta': ['F1-Score: Lista Verde', 'F1-Score: Lista Roja', 'F1-Score: Lista Verde', 'F1-Score: Lista Roja'],
    'Rendimiento': [
        reporte_pre['LISTA VERDE']['f1-score'], reporte_pre['LISTA ROJA']['f1-score'],
        reporte_post['LISTA VERDE']['f1-score'], reporte_post['LISTA ROJA']['f1-score']
    ]
}
df_metricas = pd.DataFrame(dicc_metricas)

plt.figure(figsize=(9, 5))
ax = sns.barplot(data=df_metricas, x='Clase de Alerta', y='Rendimiento', hue='Modelo Evaluado', palette='Set2')
plt.ylim(0, 1.15)
plt.title("Justificación del Proyecto: Rendimiento F1-Score Antes vs Después", weight='bold')
plt.ylabel("Puntuación de Eficacia (0.00 - 1.00)")
for p in ax.patches:
    if p.get_height() > 0:
        ax.annotate(f"{p.get_height():.2f}", (p.get_x() + p.get_width() / 2., p.get_height() + 0.02), ha='center', va='center')
plt.tight_layout()
plt.show()
```

```python
# 5.4 GRÁFICO 3: Matriz de Confusión Ex-Post
cm = confusion_matrix(y_test, y_pred_post)
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=["LISTA VERDE", "LISTA ROJA"], yticklabels=["LISTA VERDE", "LISTA ROJA"])
plt.title("Auditoría Final del Agente: Matriz de Confusión Post-Implementación", weight='bold')
plt.xlabel("Predicción del Agente Inteligente")
plt.ylabel("Estado Real en Campo (OSINFOR)")
plt.tight_layout()
plt.show()
```

```python
# =========================================================================
# BLOQUE 5.5: GENERACIÓN DE CURVAS CIENTÍFICAS (LEARNING CURVE & ROC)
# =========================================================================
from sklearn.model_selection import learning_curve
from sklearn.metrics import roc_curve, auc

print("[PROCESO] Generando Curva de Aprendizaje y Curva ROC para la sección de resultados...")

# --- 1. CÁLCULO Y GRAFICACIÓN DE LA CURVA DE APRENDIZAJE ---
train_sizes, train_scores, test_scores = learning_curve(
    modelo_post, X_train, y_train, cv=5, n_jobs=-1,
    train_sizes=np.linspace(0.1, 1.0, 5), scoring='f1_macro'
)

train_mean = np.mean(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)

plt.figure(figsize=(7, 5))
plt.plot(train_sizes, train_mean, 'o-', color="r", label="Entrenamiento (Train Score)")
plt.plot(train_sizes, test_mean, 'o-', color="g", label="Validación Cruzada (Validation Score)")
plt.title("Curva de Aprendizaje del Agente (F1-Macro)", weight='bold')
plt.xlabel("Tamaño del Conjunto de Entrenamiento (Registros)")
plt.ylabel("Puntuación de Rendimiento")
plt.legend(loc="best")
plt.grid(True, linestyle='--')
plt.tight_layout()
plt.show()
```

```python
# --- 2. CÁLCULO Y GRAFICACIÓN DE LA CURVA ROC ---
y_probs = modelo_post.predict_proba(X_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_probs)
valor_auc = auc(fpr, tpr)

plt.figure(figsize=(6, 5))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'Curva ROC (AUC = {valor_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Clasificador Aleatorio (AUC = 0.50)')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.title("Curva ROC - Capacidad de Discriminación del Agente", weight='bold')
plt.xlabel("Tasa de Falsos Positivos (1 - Especificidad)")
plt.ylabel("Tasa de Verdaderos Positivos (Sensibilidad)")
plt.legend(loc="lower right")
plt.grid(True, linestyle='--')
plt.tight_layout()
plt.show()

print("[ÉXITO] Curvas analíticas exportadas correctamente.")
```

```python
# =========================================================================
# PASO A: EVALUACIÓN DE GENERALIZACIÓN (TRAIN VS TEST)
# =========================================================================
from sklearn.metrics import accuracy_score, f1_score

# 1. Predicciones en ambos conjuntos
y_pred_train = modelo_post.predict(X_train)
y_pred_test = modelo_post.predict(X_test)

# 2. Cálculo de métricas bajo el enfoque macro (sensible al desbalanceo de OSINFOR)
f1_train = f1_score(y_train, y_pred_train, average='macro')
f1_test = f1_score(y_test, y_pred_test, average='macro')

accuracy_train = accuracy_score(y_train, y_pred_train)
accuracy_test = accuracy_score(y_test, y_pred_test)

# 3. Impresión analítica de resultados para el capítulo de hallazgos
print("========================================================")
print("   AUDITORÍA METODOLÓGICA CONTRA OVERFITTING (TESIS)   ")
print("========================================================")
print(f"Métrica F1-Macro [Entrenamiento]: {f1_train:.4f}")
print(f"Métrica F1-Macro [Prueba/Test]:   {f1_test:.4f}")
print(f"Brecha de Desajuste (F1):        {abs(f1_train - f1_test):.4f}")
print("--------------------------------------------------------")
print(f"Exactitud (Accuracy) [Train]:     {accuracy_train:.4f}")
print(f"Exactitud (Accuracy) [Test]:      {accuracy_test:.4f}")
print(f"Brecha de Desajuste (Accuracy):  {abs(accuracy_train - accuracy_test):.4f}")
print("========================================================")

# Criterio de aceptación de la hipótesis
if abs(f1_train - f1_test) < 0.05:
    print("ESTADO: EXITOSO. El agente generaliza correctamente (No hay Overfitting).")
elif f1_train > 0.95 and f1_test < 0.70:
    print("ESTADO: ALERTA. El modelo presenta Overfitting (Memorización de patrones).")
else:
    print("ESTADO: Regularización aceptable, pero requiere observar la curva de aprendizaje.")
```

```python
# =========================================================================
# PASO B: VALIDACIÓN CRUZADA ESTADÍSTICA (K-FOLD CV)
# =========================================================================
from sklearn.model_selection import cross_val_score

print("\n[PROCESO] Ejecutando Validación Cruzada de 5 pliegues (K=5)...")

# Medimos el rendimiento F1-macro en 5 subconjuntos diferentes del dataset
puntuaciones_cv = cross_val_score(modelo_post, X, y, cv=5, scoring='f1_macro', n_jobs=-1)

print("--------------------------------------------------------")
print(f"Resultados por iteración (K-Fold): {puntuaciones_cv}")
print(f"Rendimiento Promedio Esperado:     {puntuaciones_cv.mean():.4f}")
print(f"Desviación Estándar (Varianza):    {puntuaciones_cv.std():.4f}")
print("========================================================")

if puntuaciones_cv.std() < 0.03:
    print("CONCLUSIÓN: El agente es altamente estable ante variaciones del entorno de datos.")
else:
    print("CONCLUSIÓN: Alta varianza detectada. El modelo depende mucho de cómo se dividen los datos.")
```

```python
# =========================================================================
# PASO C: GRÁFICO AUDITOR DE VALIDACIÓN (TESIS)
# =========================================================================
categorías = ['F1-Macro', 'Exactitud (Accuracy)']
valores_train = [f1_train, accuracy_train]
valores_test = [f1_test, accuracy_test]

x_indices = np.arange(len(categorías))
ancho_barra = 0.35

plt.figure(figsize=(8, 5))
plt.bar(x_indices - ancho_barra/2, valores_train, ancho_barra, label='Entrenamiento (Train)', color='#4c72b0')
plt.bar(x_indices + ancho_barra/2, valores_test, ancho_barra, label='Prueba (Test)', color='#c44e52')

plt.title("Control de Calidad del Agente: Evaluación de Overfitting", weight='bold', pad=15)
plt.xticks(x_indices, categorías)
plt.ylabel("Puntuación Obtenida")
plt.ylim(0, 1.15)
plt.legend(loc='lower left')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Añadir valores numéricos sobre las barras para mayor precisión de lectura
for i in range(len(categorías)):
    plt.text(i - ancho_barra/2, valores_train[i] + 0.02, f"{valores_train[i]:.3f}", ha='center', weight='bold')
    plt.text(i + ancho_barra/2, valores_test[i] + 0.02, f"{valores_test[i]:.3f}", ha='center', weight='bold')

plt.tight_layout()
plt.show()
```
