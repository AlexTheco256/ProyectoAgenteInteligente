```python
# Configuración de las Librerías
# Componentes esenciales para manipulación de datos y álgebra lineal
import pandas as pd
import numpy as np

# Herramientas de visualización estadística y gráficos
import matplotlib.pyplot as plt
import seaborn as sns

# Algoritmos de Machine Learning y métricas de evaluación de Inteligencia Artificial
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report

print("Librerías cargadas con éxito. Entorno configurado.")
```

```python
#Bloque 2: Ingesta de Datos (Sensores del Agente)
# URL del repositorio remoto en versión Raw
url_github = "https://raw.githubusercontent.com/AlexTheco256/ProyectoAgenteInteligente/main/Dataset.csv"

# Descarga y lectura con codificación adecuada para el contexto peruano (tildes y caracteres)
df_original = pd.read_csv(url_github, sep=",", encoding="latin1")

print(f"¡Conexión Exitosa! Registros detectados: {df_original.shape[0]} filas | Columnas: {df_original.shape[1]}")
print("\n--- VISTA PREVIA DE LOS PRIMEROS 5 REGISTROS DE LA DATA ORIGINAL ---")

# Mostramos la tabla interactiva de avance
df_original.head()
```

```python
#Ingeniería de Características Ex-Ante (Filtro Predictivo)

# Selección estricta de variables conocidas ANTES de viajar a campo
variables_numericas = ["AREA_TH", "AREA_PMF", "NUM_ARBOLES_APROBADOS", "CANT_ESPECIES_APROBADOS", "VOL_APROBADO"]
variables_categoricas = ["DEPARTAMENTO", "MODALIDAD_APROVECHAMIENTO"]
objetivo = "ESTADO_OBSERVATORIO"

# Filtrado y aislamiento de ruido administrativo
df_filtrado = df_original[variables_numericas + variables_categoricas + [objetivo]].copy()

# Normalización de textos numéricos y eliminación de comas de millar
for columna in variables_numericas:
    df_filtrado[columna] = df_filtrado[columna].astype(str).str.replace(",", "", regex=False).str.strip()
    df_filtrado[columna] = pd.to_numeric(df_filtrado[columna], errors='coerce')

# Remoción de expedientes incompletos o nulos
df_filtrado = df_filtrado.dropna()

print(f"Limpieza completada. Registros útiles procesados: {df_filtrado.shape[0]} filas.")
```

```python
#Reporte Estadístico del Inventario Forestal
print("--- RESUMEN ESTADÍSTICO DE LAS VARIABLES NUMÉRICAS SELECCIONADAS ---")
# Genera métricas clave (promedios, valores mínimos, máximos y desviaciones) de las solicitudes
df_filtrado[variables_numericas].describe()
```

```python
#Codificación Binaria del Entorno (X, y)

# Mapeo matemático de la respuesta (0 = Verde / Bajo Riesgo, 1 = Roja / Alerta de Fraude)
y = df_filtrado[objetivo].map({"LISTA VERDE": 0, "LISTA ROJA": 1})

# Aislamos las preguntas de las respuestas
X_previa = df_filtrado.drop(columns=[objetivo])

# Aplicación de One-Hot Encoding para expandir categorías de texto a columnas binarias
X = pd.get_dummies(X_previa, columns=variables_categoricas, drop_first=True)

print("--- MATRIZ FINAL PROCESADA PARA EL APRENDIZAJE SUPERVISADO ---")
print(f"Dimensiones de entrada: {X.shape[0]} filas estructuradas con {X.shape[1]} características.")
print("\nEstructura de las características codificadas (Primeros 5 registros):")

# Mostramos la tabla transformada lista para los algoritmos
X.head()
```
