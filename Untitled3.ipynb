```python
# =========================================================================
# BLOQUE 1: PREPARACIÓN DE LOS DATOS (Dataset OSINFOR)
# =========================================================================
#!pip install pandas numpy matplotlib seaborn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("[INFO] Cargando registros originales de OSINFOR...")
df = pd.read_csv("https://raw.githubusercontent.com/AlexTheco256/ProyectoAgenteInteligente/main/Datasets.csv")

# 1.1 Filtrado de características críticas (Remoción de variables que causan Data Leakage)
# Evitamos usar columnas de fiscalización ex-post como 'VOLUMEN_ILEGAL' o 'ARBOLES_INEXISTENTES'
features_validas = [
    'MODALIDAD_APROVECHAMIENTO', 'DEPARTAMENTO', 'AREA_TH', 'AREA_PMF',
    'NUM_ARBOLES_APROBADOS', 'CANT_ESPECIES_APROBADOS', 'VOL_APROBADO', 'ESTADO_OBSERVATORIO'
]
df_limpio = df[features_validas].copy()

# 1.2 Tratamiento de valores nulos (Imputación por mediana para evitar sesgos en la escala)
for col in df_limpio.select_dtypes(include=[np.number]).columns:
    df_limpio[col] = df_limpio[col].fillna(df_limpio[col].median())

# 1.3 Mapeo y codificación de variables categóricas (One-Hot Encoding)
X_raw = df_limpio.drop(columns=['ESTADO_OBSERVATORIO'])
y = df_limpio['ESTADO_OBSERVATORIO'].map({'LISTA VERDE': 0, 'LISTA ROJA': 1})

# Codificación dummy de regiones y modalidades
X = pd.get_dummies(X_raw, columns=['DEPARTAMENTO', 'MODALIDAD_APROVECHAMIENTO'], drop_first=True)

print(f"[ÉXITO] Matriz de características lista. Dimensiones: {X.shape}")
```

```python
# =========================================================================
# BLOQUE 2: DIVISIÓN DEL DATASET Y MODELO BASELINE (EX-ANTE)
# =========================================================================
#!pip install scikit-learn
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report

# 2.1 Partición balanceada y estratificada (75% Entrenamiento / 25% Test)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, stratify=y, random_state=42
)

# 2.2 Modelo Baseline: Árbol de Decisión Libre (Sin control de profundidad)
print("[PROCESO] Entrenando modelo de control (Pre-Implementación)...")
modelo_pre = DecisionTreeClassifier(max_depth=None, random_state=42)
modelo_pre.fit(X_train, y_train)

# Evaluación inicial
y_pred_pre = modelo_pre.predict(X_test)
reporte_pre = classification_report(y_test, y_pred_pre, target_names=["LISTA VERDE", "LISTA ROJA"], output_dict=True)
print("\n--- REPORTE EX-ANTE (ÁRBOL SIMPLE) ---")
print(classification_report(y_test, y_pred_pre, target_names=["LISTA VERDE", "LISTA ROJA"]))
```

```python
# =========================================================================
# BLOQUE 3 Y 4: CONFIGURACIÓN, HIPERPARÁMETROS Y GESTIÓN DE DESBALANCEO
# =========================================================================
from sklearn.ensemble import RandomForestClassifier

print("[PROCESO] Configurando Agente Inteligente Regularizado (Post-Implementación)...")

# Inicialización con parámetros que controlan la varianza (Evita memorización/Overfitting)
modelo_post = RandomForestClassifier(
    n_estimators=300,            # Robustez estadística mediante número de estimadores
    max_depth=12,                # Límite de profundidad estricto contra Overfitting
    min_samples_split=6,         # Mínimo de muestras para crear ramas nuevas
    min_samples_leaf=4,          # Estabilidad de observaciones mínimas en nodos terminales
    class_weight='balanced',     # Solución directa al desbalanceo (Bloque 4)
    max_features='sqrt',         # Selección aleatoria de variables por árbol
    random_state=42,
    n_jobs=-1
)

# Entrenamiento del ensamble optimizado
modelo_post.fit(X_train, y_train)
print("[ÉXITO] Modelo Random Forest entrenado bajo restricciones matemáticas.")
```

```python
# =========================================================================
# BLOQUE 5: EVALUACIÓN CORRECTA DEL MODELO Y EXPORTACIÓN DE GRÁFICOS
# =========================================================================
from sklearn.metrics import confusion_matrix
```

```python
# =========================================================================
# PASO A: EVALUACIÓN DE GENERALIZACIÓN (TRAIN VS TEST)
# =========================================================================
from sklearn.metrics import accuracy_score, f1_score

# 1. Predicciones en ambos conjuntos
y_pred_train = modelo_post.predict(X_train)
y_pred_test = modelo_post.predict(X_test)

# 2. Cálculo de métricas bajo el enfoque macro (sensible al desbalanceo de OSINFOR)
f1_train = f1_score(y_train, y_pred_train, average='macro')
f1_test = f1_score(y_test, y_pred_test, average='macro')

accuracy_train = accuracy_score(y_train, y_pred_train)
accuracy_test = accuracy_score(y_test, y_pred_test)

# 3. Impresión analítica de resultados para el capítulo de hallazgos
print("========================================================")
print("   AUDITORÍA METODOLÓGICA CONTRA OVERFITTING (TESIS)   ")
print("========================================================")
print(f"Métrica F1-Macro [Entrenamiento]: {f1_train:.4f}")
print(f"Métrica F1-Macro [Prueba/Test]:   {f1_test:.4f}")
print(f"Brecha de Desajuste (F1):        {abs(f1_train - f1_test):.4f}")
print("--------------------------------------------------------")
print(f"Exactitud (Accuracy) [Train]:     {accuracy_train:.4f}")
print(f"Exactitud (Accuracy) [Test]:      {accuracy_test:.4f}")
print(f"Brecha de Desajuste (Accuracy):  {abs(accuracy_train - accuracy_test):.4f}")
print("========================================================")

# Criterio de aceptación de la hipótesis
if abs(f1_train - f1_test) < 0.05:
    print("ESTADO: EXITOSO. El agente generaliza correctamente (No hay Overfitting).")
elif f1_train > 0.95 and f1_test < 0.70:
    print("ESTADO: ALERTA. El modelo presenta Overfitting (Memorización de patrones).")
else:
    print("ESTADO: Regularización aceptable, pero requiere observar la curva de aprendizaje.")
```

```python
#Bloque 2: Ingesta de Datos (Sensores del Agente)
# URL del repositorio remoto en versión Raw
url_github = "https://raw.githubusercontent.com/AlexTheco256/ProyectoAgenteInteligente/main/Dataset.csv"

# Descarga y lectura con codificación adecuada para el contexto peruano (tildes y caracteres)
df_original = pd.read_csv(url_github, sep=",", encoding="latin1")

print(f"¡Conexión Exitosa! Registros detectados: {df_original.shape[0]} filas | Columnas: {df_original.shape[1]}")
print("\n--- VISTA PREVIA DE LOS PRIMEROS 5 REGISTROS DE LA DATA ORIGINAL ---")

# Mostramos la tabla interactiva de avance
df_original.head()
```

```python
# 1. Definición de variables permitidas para evitar la filtración de datos (Data Leakage)
variables_numericas = ["AREA_TH", "AREA_PMF", "NUM_ARBOLES_APROBADOS", "CANT_ESPECIES_APROBADOS", "VOL_APROBADO"]
variables_categoricas = ["DEPARTAMENTO", "MODALIDAD_APROVECHAMIENTO"]
objetivo = "ESTADO_OBSERVATORIO"

# 2. Creamos una copia con las columnas seleccionadas
df_filtrado = df_original[variables_numericas + variables_categoricas + [objetivo]].copy()

# 3. Limpieza de formatos numéricos (conversión limpia a números)
for columna in variables_numericas:
    df_filtrado[columna] = df_filtrado[columna].astype(str).str.replace(",", "", regex=False).str.strip()
    df_filtrado[columna] = pd.to_numeric(df_filtrado[columna], errors='coerce')

# 4. Eliminación de registros con valores vacíos o inconsistentes
df_filtrado = df_filtrado.dropna()

print(f"Limpieza completada. Registros válidos para el agente: {df_filtrado.shape[0]} filas.")
print("\n--- RESUMEN ESTADÍSTICO DE LAS VARIABLES NUMÉRICAS SELECCIONADAS ---")

# Mostramos una tabla con métricas clave (promedio, desviación, min, max)
df_filtrado[variables_numericas].describe()
```

```python
# 1. Mapeo de la variable objetivo (0 = Verde, 1 = Roja)
y = df_filtrado[objetivo].map({"LISTA VERDE": 0, "LISTA ROJA": 1})

# 2. Aislamos las características de entrada descartando la respuesta
X_previa = df_filtrado.drop(columns=[objetivo])

# 3. Transformación de categorías de texto a columnas numéricas (One-Hot Encoding)
X = pd.get_dummies(X_previa, columns=variables_categoricas, drop_first=True)

print("--- ESTRUCTURA FINAL DE LAS VARIABLES DE ENTRADA (X) COMPUTADAS ---")
print(f"Matriz procesada para el agente: {X.shape[0]} registros listos con {X.shape[1]} características numéricas.")
print("\nColumnas codificadas finales (Vista de las primeras 5 filas):")

# Mostramos cómo lucen ahora las variables tras la transformación digital
X.head()
```

```python
# 1. Definición del Árbol de Decisión
def entrenar_arbol(X_train, y_train):
    # max_depth=5 evita que el árbol memorice la data (overfitting) y mantenga reglas lógicas simples
    modelo = DecisionTreeClassifier(
        max_depth=5,
        random_state=42
    )
    modelo.fit(X_train, y_train)
    return modelo

# 2. Definición del Bosque Aleatorio (Random Forest)
def entrenar_random_forest(X_train, y_train):
    # Creamos un ensamble de 100 árboles independientes que votarán por el resultado final
    modelo = RandomForestClassifier(
        n_estimators=100,
        random_state=42
    )
    modelo.fit(X_train, y_train)
    return modelo

print("Funciones de entrenamiento configuradas y listas para procesar.")
```

```python
# 1. Mostrar información básica de los registros que ingresan al modelo
print("\n===== INFORMACIÓN GENERAL DE INGRESO =====")
print(f"Cantidad de registros útiles para el agente: {len(X)}")
print("\nDistribución real de clases en la data (0 = LISTA VERDE, 1 = LISTA ROJA):")
print(y.value_counts())
```

```python
#Codificación Binaria del Entorno (X, y)

# Mapeo matemático de la respuesta (0 = Verde / Bajo Riesgo, 1 = Roja / Alerta de Fraude)
y = df_filtrado[objetivo].map({"LISTA VERDE": 0, "LISTA ROJA": 1})

# Aislamos las preguntas de las respuestas
X_previa = df_filtrado.drop(columns=[objetivo])

# Aplicación de One-Hot Encoding para expandir categorías de texto a columnas binarias
X = pd.get_dummies(X_previa, columns=variables_categoricas, drop_first=True)

print("--- MATRIZ FINAL PROCESADA PARA EL APRENDIZAJE SUPERVISADO ---")
print(f"Dimensiones de entrada: {X.shape[0]} filas estructuradas con {X.shape[1]} características.")
print("\nEstructura de las características codificadas (Primeros 5 registros):")

# Mostramos la tabla transformada lista para los algoritmos
X.head()
```

```python
#Simulación del Entorno Futuro (Partición 80/20)
# División del dataset manteniendo el balance original de infracciones (stratify)
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("===== PARTICIÓN DEL ENTORNO COMPLETADA =====")
print(f"Muestra destinada al Entrenamiento del Cerebro: {X_train.shape[0]} expedientes.")
print(f"Muestra destinada a la Evaluación de Alertas (Casos Nuevos): {X_test.shape[0]} expedientes.")
```

```python
#Gráfico 1 - Visualización de la Distribución de Clases
#Configuración y diseño del gráfico de barras
plt.figure(figsize=(6,4))
sns.countplot(x=y, palette="Set2")

plt.title("Distribución de LISTA VERDE (0) y LISTA ROJA (1)")
plt.xlabel("Clasificación Final en el Observatorio de OSINFOR")
plt.ylabel("Cantidad de Expedientes")

plt.tight_layout()
plt.savefig("grafico_distribucion_clases.png") # Exporta el archivo al panel lateral
plt.show()
```

```python
#Entrenamiento y Competencia de Modelos (Árbol vs. Bosque)
from imblearn.over_sampling import SMOTE

# Balanceo de Clases antes del entrenamiento
smote = SMOTE(random_state=42)
X_train, y_train = smote.fit_resample(X_train, y_train)

print(f"Clases balanceadas — Verde: {(y_train==0).sum()} | Roja: {(y_train==1).sum()}")

print("1. Entrenando Agente mediante Árbol de Decisión...")
# Profundidad controlada para evitar sobreajuste (memorización del ruido)
modelo_dt = DecisionTreeClassifier(max_depth=5, random_state=42)
modelo_dt.fit(X_train, y_train)
y_pred_dt = modelo_dt.predict(X_test)

print("2. Entrenando Agente mediante Bosque Aleatorio (Random Forest)...")
# Ensamble de 100 árboles con mecanismo de votación mayoritaria
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)

print("\n=======================================================")
print("REPORTE DE RENDIMIENTO - AGENTE ÁRBOL DE DECISIÓN")
print("=======================================================")
print(classification_report(y_test, y_pred_dt, target_names=["LISTA VERDE", "LISTA ROJA"]))

print("\n=======================================================")
print("REPORTE DE RENDIMIENTO - AGENTE RANDOM FOREST")
print("=======================================================")
print(classification_report(y_test, y_pred_rf, target_names=["LISTA VERDE", "LISTA ROJA"]))
```

```python
#Gráfico 2 - Matriz de Confusión (Diagnóstico Clínico del Agente)
# Computar los cuadrantes de la matriz matemática
matriz = confusion_matrix(y_test, y_pred_rf)

# Construcción del Heatmap interactivo
plt.figure(figsize=(6,5))
sns.heatmap(
    matriz,
    annot=True,
    fmt="d",
    cmap="Blues",
    xticklabels=["LISTA VERDE", "LISTA ROJA"],
    yticklabels=["LISTA VERDE", "LISTA ROJA"]
)

plt.title("Matriz de Confusión - Diagnóstico de Alerta Temprana")
plt.xlabel("Predicción Emitida por el Agente")
plt.ylabel("Estado Real Registrado en Campo")

plt.tight_layout()
plt.savefig("matriz_confusion_rf.png")
plt.show()
from IPython.display import Image, display
display(Image(filename="Matriz de Confusión.png", width=600))
```

```python
# Producción y Despliegue (El Actuador del Agente)
def predecir_nuevo_expediente(datos_nuevos, modelo, columnas_entrenamiento):
    df_nuevo = pd.DataFrame([datos_nuevos])
    df_nuevo_codificado = pd.get_dummies(df_nuevo)
    df_nuevo_final = df_nuevo_codificado.reindex(columns=columnas_entrenamiento, fill_value=0)

    prediccion = modelo.predict(df_nuevo_final)[0]
    probabilidades = modelo.predict_proba(df_nuevo_final)[0]

    estado_dictamen = "⚠️ ALERTA: ALTO RIESGO (LISTA ROJA)" if prediccion == 1 else "✅ COMPORTAMIENTO NORMAL (LISTA VERDE)"
    probabilidad_riesgo = probabilidades[1] * 100

    print("\n=======================================================")
    print("      DIAGNÓSTICO EMITIDO POR EL AGENTE INTELIGENTE")
    print("=======================================================")
    print(f"Dictamen del Agente: {estado_dictamen}")
    print(f"Probabilidad de Infracción o Fraude: {probabilidad_riesgo:.2f}%")
    print("Recomendación Operativa: " + ("PRIORIZAR INSPECCIÓN PRESENCIAL INMEDIATA" if prediccion == 1 else "SUPERVISIÓN ESTÁNDAR PROGRAMADA"))
    print("=======================================================")

# --- CASO DE SIMULACIÓN INTERACTIVO ---
expediente_ejemplo = {
    "AREA_TH": 1200.50,
    "AREA_PMF": 150.20,
    "NUM_ARBOLES_APROBADOS": 450,
    "CANT_ESPECIES_APROBADOS": 12,
    "VOL_APROBADO": 3800.75,
    "DEPARTAMENTO": "LORETO",
    "MODALIDAD_APROVECHAMIENTO": "PREDIO PRIVADO"
}

# Ejecución del actuador del agente
predecir_nuevo_expediente(expediente_ejemplo, rf, X.columns)
```
