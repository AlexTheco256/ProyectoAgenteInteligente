```python
#Bloque 2: Ingesta de Datos (Sensores del Agente)
# URL del repositorio remoto en versión Raw
url_github = "https://raw.githubusercontent.com/AlexTheco256/ProyectoAgenteInteligente/main/Dataset.csv"

# Descarga y lectura con codificación adecuada para el contexto peruano (tildes y caracteres)
df_original = pd.read_csv(url_github, sep=",", encoding="latin1")

print(f"¡Conexión Exitosa! Registros detectados: {df_original.shape[0]} filas | Columnas: {df_original.shape[1]}")
print("\n--- VISTA PREVIA DE LOS PRIMEROS 5 REGISTROS DE LA DATA ORIGINAL ---")

# Mostramos la tabla interactiva de avance
df_original.head()
```

```python
# 1. Definición de variables permitidas para evitar la filtración de datos (Data Leakage)
variables_numericas = ["AREA_TH", "AREA_PMF", "NUM_ARBOLES_APROBADOS", "CANT_ESPECIES_APROBADOS", "VOL_APROBADO"]
variables_categoricas = ["DEPARTAMENTO", "MODALIDAD_APROVECHAMIENTO"]
objetivo = "ESTADO_OBSERVATORIO"

# 2. Creamos una copia con las columnas seleccionadas
df_filtrado = df_original[variables_numericas + variables_categoricas + [objetivo]].copy()

# 3. Limpieza de formatos numéricos (conversión limpia a números)
for columna in variables_numericas:
    df_filtrado[columna] = df_filtrado[columna].astype(str).str.replace(",", "", regex=False).str.strip()
    df_filtrado[columna] = pd.to_numeric(df_filtrado[columna], errors='coerce')

# 4. Eliminación de registros con valores vacíos o inconsistentes
df_filtrado = df_filtrado.dropna()

print(f"Limpieza completada. Registros válidos para el agente: {df_filtrado.shape[0]} filas.")
print("\n--- RESUMEN ESTADÍSTICO DE LAS VARIABLES NUMÉRICAS SELECCIONADAS ---")

# Mostramos una tabla con métricas clave (promedio, desviación, min, max)
df_filtrado[variables_numericas].describe()
```

```python
# 1. Mapeo de la variable objetivo (0 = Verde, 1 = Roja)
y = df_filtrado[objetivo].map({"LISTA VERDE": 0, "LISTA ROJA": 1})

# 2. Aislamos las características de entrada descartando la respuesta
X_previa = df_filtrado.drop(columns=[objetivo])

# 3. Transformación de categorías de texto a columnas numéricas (One-Hot Encoding)
X = pd.get_dummies(X_previa, columns=variables_categoricas, drop_first=True)

print("--- ESTRUCTURA FINAL DE LAS VARIABLES DE ENTRADA (X) COMPUTADAS ---")
print(f"Matriz procesada para el agente: {X.shape[0]} registros listos con {X.shape[1]} características numéricas.")
print("\nColumnas codificadas finales (Vista de las primeras 5 filas):")

# Mostramos cómo lucen ahora las variables tras la transformación digital
X.head()
```

```python
# 1. Definición del Árbol de Decisión
def entrenar_arbol(X_train, y_train):
    # max_depth=5 evita que el árbol memorice la data (overfitting) y mantenga reglas lógicas simples
    modelo = DecisionTreeClassifier(
        max_depth=5,
        random_state=42
    )
    modelo.fit(X_train, y_train)
    return modelo

# 2. Definición del Bosque Aleatorio (Random Forest)
def entrenar_random_forest(X_train, y_train):
    # Creamos un ensamble de 100 árboles independientes que votarán por el resultado final
    modelo = RandomForestClassifier(
        n_estimators=100,
        random_state=42
    )
    modelo.fit(X_train, y_train)
    return modelo

print("Funciones de entrenamiento configuradas y listas para procesar.")
```

```python
# 1. Mostrar información básica de los registros que ingresan al modelo
print("\n===== INFORMACIÓN GENERAL DE INGRESO =====")
print(f"Cantidad de registros útiles para el agente: {len(X)}")
print("\nDistribución real de clases en la data (0 = LISTA VERDE, 1 = LISTA ROJA):")
print(y.value_counts())
```

```python
#Codificación Binaria del Entorno (X, y)

# Mapeo matemático de la respuesta (0 = Verde / Bajo Riesgo, 1 = Roja / Alerta de Fraude)
y = df_filtrado[objetivo].map({"LISTA VERDE": 0, "LISTA ROJA": 1})

# Aislamos las preguntas de las respuestas
X_previa = df_filtrado.drop(columns=[objetivo])

# Aplicación de One-Hot Encoding para expandir categorías de texto a columnas binarias
X = pd.get_dummies(X_previa, columns=variables_categoricas, drop_first=True)

print("--- MATRIZ FINAL PROCESADA PARA EL APRENDIZAJE SUPERVISADO ---")
print(f"Dimensiones de entrada: {X.shape[0]} filas estructuradas con {X.shape[1]} características.")
print("\nEstructura de las características codificadas (Primeros 5 registros):")

# Mostramos la tabla transformada lista para los algoritmos
X.head()
```

```python
#Simulación del Entorno Futuro (Partición 80/20)
# División del dataset manteniendo el balance original de infracciones (stratify)
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("===== PARTICIÓN DEL ENTORNO COMPLETADA =====")
print(f"Muestra destinada al Entrenamiento del Cerebro: {X_train.shape[0]} expedientes.")
print(f"Muestra destinada a la Evaluación de Alertas (Casos Nuevos): {X_test.shape[0]} expedientes.")
```

```python
#Gráfico 1 - Visualización de la Distribución de Clases
#Configuración y diseño del gráfico de barras
plt.figure(figsize=(6,4))
sns.countplot(x=y, palette="Set2")

plt.title("Distribución de LISTA VERDE (0) y LISTA ROJA (1)")
plt.xlabel("Clasificación Final en el Observatorio de OSINFOR")
plt.ylabel("Cantidad de Expedientes")

plt.tight_layout()
plt.savefig("grafico_distribucion_clases.png") # Exporta el archivo al panel lateral
plt.show()
```

```python
#Entrenamiento y Competencia de Modelos (Árbol vs. Bosque)
from imblearn.over_sampling import SMOTE

# Balanceo de Clases antes del entrenamiento
smote = SMOTE(random_state=42)
X_train, y_train = smote.fit_resample(X_train, y_train)

print(f"Clases balanceadas — Verde: {(y_train==0).sum()} | Roja: {(y_train==1).sum()}")

print("1. Entrenando Agente mediante Árbol de Decisión...")
# Profundidad controlada para evitar sobreajuste (memorización del ruido)
modelo_dt = DecisionTreeClassifier(max_depth=5, random_state=42)
modelo_dt.fit(X_train, y_train)
y_pred_dt = modelo_dt.predict(X_test)

print("2. Entrenando Agente mediante Bosque Aleatorio (Random Forest)...")
# Ensamble de 100 árboles con mecanismo de votación mayoritaria
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)

print("\n=======================================================")
print("REPORTE DE RENDIMIENTO - AGENTE ÁRBOL DE DECISIÓN")
print("=======================================================")
print(classification_report(y_test, y_pred_dt, target_names=["LISTA VERDE", "LISTA ROJA"]))

print("\n=======================================================")
print("REPORTE DE RENDIMIENTO - AGENTE RANDOM FOREST")
print("=======================================================")
print(classification_report(y_test, y_pred_rf, target_names=["LISTA VERDE", "LISTA ROJA"]))
```
